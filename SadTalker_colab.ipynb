```python
# 1. Очистка старой установки
!rm -rf /content/SadTalker

# 2. Клонирование официального репозитория
%cd /content
!git clone https://github.com/Winfredy/SadTalker.git

# 3. Переход в рабочую директорию
%cd /content/SadTalker

# 4. Установка совместимых версий библиотек
# Мы принудительно ставим numpy версии 1.26.4, так как в 2.0 удалены нужные атрибуты
!pip install "numpy<2.0.0"
!pip install -r requirements.txt
# Фиксируем версию Gradio для стабильной работы веб-интерфейса
!pip install "gradio<4.0.0"

# 5. Загрузка моделей (checkpoints) с использованием aria2 для высокой скорости
!apt-get install -y aria2

# Загрузка маппинг-моделей
!aria2c --console-log-level=error -c -x 16 -s 16 -k 1M https://github.com/OpenTalker/SadTalker/releases/download/v0.0.2-rc/mapping_00109-model.pth.tar -d checkpoints -o mapping_00109-model.pth.tar
!aria2c --console-log-level=error -c -x 16 -s 16 -k 1M https://github.com/OpenTalker/SadTalker/releases/download/v0.0.2-rc/mapping_00229-model.pth.tar -d checkpoints -o mapping_00229-model.pth.tar

# Загрузка основных весов SadTalker
!aria2c --console-log-level=error -c -x 16 -s 16 -k 1M https://github.com/OpenTalker/SadTalker/releases/download/v0.0.2-rc/SadTalker_V0.0.2_256.safetensors -d checkpoints -o SadTalker_V0.0.2_256.safetensors
!aria2c --console-log-level=error -c -x 16 -s 16 -k 1M https://github.com/OpenTalker/SadTalker/releases/download/v0.0.2-rc/SadTalker_V0.0.2_512.safetensors -d checkpoints -o SadTalker_V0.0.2_512.safetensors

# Загрузка весов для улучшения лица (GFPGAN)
!aria2c --console-log-level=error -c -x 16 -s 16 -k 1M https://github.com/TencentARC/GFPGAN/releases/download/v1.3.0/GFPGANv1.4.pth -d gfpgan/weights -o GFPGANv1.4.pth
!aria2c --console-log-level=error -c -x 16 -s 16 -k 1M https://github.com/OpenTalker/SadTalker/releases/download/v0.0.2-rc/alignment_WFLW_4HG.pth -d gfpgan/weights -o alignment_WFLW_4HG.pth
!aria2c --console-log-level=error -c -x 16 -s 16 -k 1M https://github.com/OpenTalker/SadTalker/releases/download/v0.0.2-rc/detection_Resnet50_Final.pth -d gfpgan/weights -o detection_Resnet50_Final.pth
!aria2c --console-log-level=error -c -x 16 -s 16 -k 1M https://github.com/OpenTalker/SadTalker/releases/download/v0.0.2-rc/parsing_parsenet.pth -d gfpgan/weights -o parsing_parsenet.pth

# 6. Запуск веб-интерфейса
# После выполнения появится ссылка "public URL", по которой можно открыть приложение
!python app_sadtalker.py --share
```
